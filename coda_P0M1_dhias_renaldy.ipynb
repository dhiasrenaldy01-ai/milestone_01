## A. Extract

````python
nama_clean = []
brand_clean = []
harga_clean = []
terjual_clean = []
gambar_clean = []


driver = webdriver.Chrome()
for hal in range(1,7):
    print(f'processing halaman {hal}')
    url = f'https://www.ebay.com/b/Laptops-Netbooks/175672/bn_1648276?_pgn={hal}'


    driver.get(url)
    time.sleep(6)
    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')

    boxes = soup.find_all('div',{'class':'brwrvr__item-card__wrapper'})

    for box in boxes:

        # Nama Produk
        try:
            el_nama = box.find('h3',{'class':'textual-display bsig__title__text'})
            nama_clean.append(el_nama.get_text())
        except:
            nama_clean.append(None)
            
        # Brand
        try:
            el_brand = box.find('span',{'class':'textual-display bsig__generic bsig__listingCondition secondary'})
            brand_clean.append(el_brand.get_text())
        except:
            brand_clean.append(None)

        # Harga
        try:
            el_harga = box.find('span',{'class':'textual-display bsig__price bsig__price--displayprice'})
            harga_clean.append(el_harga.get_text())
        except:
            harga_clean.append(None)
             
        # Terjual
        try:
            el_terjual = box.find('span',{'class':'textual-display negative'})
            terjual_clean.append(el_terjual.get_text())
        except:
            terjual_clean.append(None)

        # Gambar
        try:
            el_gambar = box.find('img',{'class':'brwrvr__item-card__image'})
            gambar_clean.append(el_gambar['src'])
        except:
            gambar_clean.append(None)

        print(f'finish halaman {hal}')

df_data = pd.DataFrame()
df_data['nama'] = nama_clean
df_data['brand'] = brand_clean
df_data['harga'] = harga_clean
df_data['terjual'] = terjual_clean
df_data['gambar'] = gambar_clean

driver.quit()

print('Process Success')
````

## 3. Membersihkan Data Numerik

````python
# -----------------------------

# Menghapus baris yang mengandung simbol dollar "$" pada kolom harga ---
mask_dollar = df['harga'].astype(str).str.contains(r'\$', regex=True)
df_clean = df[~mask_dollar].copy()

print(f"\nJumlah baris dihapus (mengandung $): {mask_dollar.sum()}")

# Membersihkan karakter selain angka pada kolom numerik ---
# Contoh: harga "Rp 1.200.000" ‚Üí 1200000

def to_numeric(series):
    return (series.astype(str)
                  .str.replace(r'[^0-9]', '', regex=True)
                  .replace('', '0')
                  .astype(int))

# Hapus data yang mengandung simbol '$'
df = df[~df['harga'].astype(str).str.contains(r'\$', regex=True)]

# BERSIHKAN KOLOM HARGA

# === 1. Ambil harga pertama jika ada "to" ===
df['harga'] = df['harga'].astype(str).str.split(' to ').str[0]

# === 2. Hapus IDR, koma, titik, spasi, dan karakter selain angka ===
df['harga'] = (
    df['harga']
    .str.replace(r'IDR', '', regex=True)
    .str.replace(r'[^0-9]', '', regex=True)     # hapus semua selain angka
)

# === 3. Jika kosong ‚Üí isi 0 ===
df['harga'] = df['harga'].replace('', '0')

# === 4. Ubah ke integer ===
df['harga'] = df['harga'].astype(int)

# == 5. Ubah NaN jadi None
df = df.replace({pd.NA: None, pd.NaT: None, float('nan'): None})

# BERSIHKAN KOLOM TERJUAL

# --- 1. Ubah NaN menjadi None ---
df['terjual'] = df['terjual'].where(pd.notnull(df['terjual']), None)

# --- 2. Ambil angka saja dari string ---
df['terjual'] = (
    df['terjual']
        .astype(str)
        .str.extract(r'(\d+)')       # ambil angka pertama
)

# --- 3. Jika kosong ‚Üí ganti 0 ---
df['terjual'] = df['terjual'].replace({None: '0', 'None': '0', 'nan': '0'})

# --- 4. Hapus koma & ubah ke integer ---
df['terjual'] = (
    df['terjual']
    .str.replace(',', '', regex=False)
    .astype(int)
)

df
````
